```python
%matplotlib inline

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

# define an affine subspace of R^2

```python
x=np.array([100, 6])
U=np.array([7, 20])
line = np.linspace(25, -25, 1000)
Vx = []
Vy = []

for i in line:
  p = x + np.dot(i, U)
  plt.scatter(p[0], p[1], s = 10, c = 'red')
  Vx.append(p[0])
  Vy.append(p[1])

  

V = np.array([Vx, Vy]).T
```

# A is a transformation matrix and p is  a translation matrix for affine mapping

```python
A=np.array([[4, -2/5],
            [-10, -1/3]])
p=np.array([25, -25])
Ux = []
Uy = []


for i in range(V.shape[0]):
  pp = p + np.dot(A, V[i])
  plt.scatter(pp[0], pp[1], s = 10, c = 'green')
  Ux.append(pp[0])
  Uy.append(pp[1])

  
U= np.array([Ux, Uy]).T
```

# compare tow of them

```python
plt.axvline(0, c='black')
plt.axhline(0, c='black')
plt.scatter(Vx, Vy, s = 10, c = 'red')
plt.scatter(Ux, Uy, s = 10, c = 'green')
```
