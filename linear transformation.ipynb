```python
import matplotlib.pyplot as plt
import numpy as np

# 將參數設定好
a, b = 0, 1
c, d = -1, 2

# [0, 1] 等分 50 個點，[-1, 2] 等分 50 個點
n, m = 50, 50
x = np.linspace(a, b, n)
y = np.linspace(c, d, n)
```

```python
xx, yy = np.meshgrid(x, y)
xx, yy = xx.flatten(), yy.flatten()
S_meshgrid = np.vstack([xx, yy])

# 逐點畫出
plt.scatter(xx, yy)
plt.title('$S$')

# 改變圖片的呈現範圍
plt.xlim(-1.5, 2)
plt.ylim(-2, 3)
```

```python
##水平拉長3倍
A = np.array([[3, 0],
              [0, 1]])

def phi(x):
    assert A.shape[1]==x.shape[0]
    return A@x
```

```python
T = phi(S_meshgrid)
u, v = phi(S_meshgrid)

# 畫出 phi(S)
plt.scatter(u, v, c='r')
plt.title('$\Phi(S)$')
```

```python
##拉長後在旋轉pi/3
A = np.array([[np.cos(np.pi/3), -np.sin(np.pi/3)],
              [np.sin(np.pi/3), np.cos(np.pi/3)]])
```

```python
T = phi(S_meshgrid)
u, v = phi(S_meshgrid)

# 畫出 phi(S)
plt.scatter(u, v, c='r')
plt.title('$\Phi(S)$')
```
