```python
import matplotlib.pyplot as plt
import numpy as np

# 將參數設定好
a, b = 0, 1
c, d = -1, 2

# [0, 1] 等分 50 個點，[-1, 2] 等分 50 個點
n, m = 50, 50
x = np.linspace(a, b, n)
y = np.linspace(c, d, n)
```

```python
xx, yy = np.meshgrid(x, y)
xx, yy = xx.flatten(), yy.flatten()
S_meshgrid = np.vstack([xx, yy])

# 逐點畫出
plt.scatter(xx, yy)
plt.title('$S$')

# 改變圖片的呈現範圍
plt.xlim(-1.5, 2)
plt.ylim(-2, 3)
```

```python
##水平拉長3倍
A = np.array([[3, 0],
              [0, 1]])

def phi(x):
    assert A.shape[1]==x.shape[0]
    return A@x
```

```python
T = phi(S_meshgrid)
u, v = phi(S_meshgrid)

# 畫出 phi(S)
plt.scatter(u, v, c='r')
plt.title('$\Phi(S)$')
```

```python
##拉長後在旋轉pi/3
A = np.array([[np.cos(np.pi/3), -np.sin(np.pi/3)],
              [np.sin(np.pi/3), np.cos(np.pi/3)]])
```

```python
T = phi(S_meshgrid)
u, v = phi(S_meshgrid)

# 畫出 phi(S)
plt.scatter(u, v, c='r')
plt.title('$\Phi(S)$')
```

```python
#旋轉後再斜拉
A = np.array([[3, -1/2],
              [1/3, 1/4]])
```

```python
T = phi(S_meshgrid)
u, v = phi(S_meshgrid)

# 畫出 phi(S)

plt.scatter(xx, yy, alpha=0.1)
plt.scatter(u, v, c='r')
plt.title('$S$ and $\Phi(S)$')

# 改變圖片的呈現範圍
plt.xlim(-2, 4)
plt.ylim(-2, 3)
```

# 橢圓

```python
p, q = 30, 60
r = np.linspace(0, 1, p)
t = np.linspace(0, 2*np.pi, q)

rr, tt = np.meshgrid(r, t)
rc, rs = 5*rr*np.cos(tt), 3*rr*np.sin(tt)

xx, yy = rc.flatten(), rs.flatten()
S_meshgrid = np.vstack([xx, yy])

# 以半徑當作顏色指標 c=-rr，半徑越大，顏色越深
plt.scatter(rc, rs, c=-rr)
plt.title('$S$')

# 改變圖片的呈現範圍
plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

```python
##函數：（3x - 1/3y, 1/4x + 1/3y）
A = np.array([[3, -1/3],
              [1/4, 1/3]])

def phi(x):
    assert A.shape[1]==x.shape[0], 'incompatible dimension'
    return A@x
```

```python
T = phi(S_meshgrid)
u, v = phi(S_meshgrid)

# 畫出 phi(S)
plt.scatter(u, v, c='r')
plt.title('$\Phi(S)$')
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

```python
plt.scatter(xx, yy, alpha=0.1)
plt.scatter(u, v, c='r')
plt.title('$S$ and $\Phi(S)$')

# 改變圖片的呈現範圍
plt.xlim(-5, 5)
plt.ylim(-5, 5)
```
